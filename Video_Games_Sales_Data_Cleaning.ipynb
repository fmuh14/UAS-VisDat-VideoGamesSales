```python
import pandas as pd

# Load the dataset
df = pd.read_csv('video_games_sales.csv',header=0)

df
```

```python
df.dtypes
```

```python
# Convert kolom 'year' ke tipe data integer
df['year'] = df['year'].astype('Int64')

df.dtypes
```

```python
# Cek nilai hilang
missing_values = df.isnull().sum()
missing_values
```

```python
df.dropna(inplace=True)  # Hapus baris dengan nilai-nilai yang hilang

missing_values = df.isnull().sum()
missing_values
```

```python
#Cek Row Duplicate
duplicate_rows = df[df.duplicated()]
duplicate_rows
```

```python
df["year"].value_counts()
```

```python
#Hapus data video game pada tahun 2017 dan 2020 karena nilainya terlalu sedikit
df = df[(df['year'] != 2017) & (df['year'] != 2020)]

df["year"].value_counts()
```

```python
df.to_csv('video_games_sales_clean.csv', index=False)
```
